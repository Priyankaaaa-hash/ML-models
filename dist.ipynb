```python
import pandas as pd
dict={
     "dist" : [50,100,150,200,250],
     "fuel" : [4,8,12,16,20]
}
df=pd.DataFrame(dict)
print(df)
```

```python
x=df[["dist"]]
y=df[["fuel"]]
from sklearn.linear_model import LinearRegression
mod =LinearRegression()
mod.fit(x,y)
cal = mod.predict([[350]])
print(cal)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,mod.predict(x))
plt.show()
```
