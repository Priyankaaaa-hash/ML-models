```python
import pandas as pd
dict={
    "ex_dur" : [2,2.5,4.7,6],
    "calories_burned" :[500,650,735,850]
}
df = pd.DataFrame(dict)
print(df)
```

```python
x=df[["ex_dur"]]
y=df[["calories_burned"]]
from sklearn.linear_model import LinearRegression
mod = LinearRegression()
mod.fit(x,y)
cal = mod.predict([[7.5]])
print(cal)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,mod.predict(x))
plt.show()
```
