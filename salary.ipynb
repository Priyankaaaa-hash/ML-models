```python
import pandas as pd
dict={
    "experience" : [1,3,4,7],
    "salary" : [3,5,7,10]
}
df=pd.DataFrame(dict)
print(df)
```

```python
x=df[["experience"]]
y=df[["salary"]]
from sklearn.linear_model import LinearRegression
sal = LinearRegression()
sal.fit(x,y)
predict = sal.predict([[8]])
print(predict)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,sal.predict(x))
plt.show()
```
