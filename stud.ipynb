```python
import pandas as pd
dict={
    "hours" : [2,4,6,8],
    "marks" : [40,60,80,100]
}
df=pd.DataFrame(dict)
print(df)
```

```python
x=df[["hours"]]
y=df[["marks"]]
from sklearn.linear_model import LinearRegression
model2=LinearRegression()
model2.fit(x,y)
predict = model2.predict([[5]])
print(predict)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,model2.predict(x))
plt.show()
```
