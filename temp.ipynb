```python
import pandas as pd
dict={
    "temp" : [20,22,25,28,30,35],
    "ice-cream-sales" :[100,120,150,180,210,280]
}
df=pd.DataFrame(dict)
print(df)
```

```python
x=df[["temp"]]
y=df[["ice-cream-sales"]]
from sklearn.linear_model import LinearRegression
mod=LinearRegression()
mod.fit(x,y)
cal = mod.predict([[45]])
print(cal)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,mod.predict(x))
plt.show()
```
